```python
##------Dot Product Example(nut butter sales)
```

```python
np.random.seed(0)
#Number of jars sold
sales_amounts = np.random.randint(20, size=(5,3))
sales_amounts
```

```python
#Create Weekly_sales Datframe
weeky_sales=pd.DataFrame(sales_amounts,index=["Mon","Tue","Wed","Thu","Fri"],columns=["Almond Butter","Peanut Butter","Cashew Butter"])
weeky_sales
```

```python
#create Pirce array
prices=np.array([10,8,12])

#Create Butter Prices DataFrame
butter_prices = pd.DataFrame(prices.reshape(1,3),index=["Price"],columns=[["Almond Butter","Peanut Butter","Cashew Butter"]])
butter_prices
```

```python
#Shape arent alighed lets transpose
total_sales=prices.dot(sales_amounts.T)
total_sales
```

```python
#Create daily sales
butter_prices
```

```python
weeky_sales["Daily-Sales"]=total_sales
```

```python
weeky_sales=weeky_sales.drop("Daily-Sales",axis=1)
```

```python
weeky_sales.shape , butter_prices.shape
```

```python
butter_prices.dot(weeky_sales.T)
```
