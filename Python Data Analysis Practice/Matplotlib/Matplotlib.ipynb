```python
#----------Two options for subplots
#Option 1
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))

#Plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(x);
```

```python
#SUbplots Option2
```

```python
fig,ax=plt.subplots(nrows=2,
                    ncols=2,
                    figsize=(10,5))

#Plot to each different index

ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(x);
```

#### Which one to use (pyplot vs matplot 00 method)
* When plotting somethig quickly use pyplot
* WHen plotting something advance then use matplot lib

```python
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
#Plypot method
over_50.plot(kind="scatter",x="age",y="chol",c="target");
```

```python
#OO method
fix , ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
#ax.set_xlim([45,100])
```

```python
over_50.target.values
```

```python
#Recreating from oo method from scratch
```

```python
fig,ax =plt.subplots(figsize=(10,6))

#plot the data

scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])
#Customize the plot
ax.set(title="Heart Disease and Cholesterol levels",
       xlabel="Age",
       ylabel="Cholestrol");

#Add a legend

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

#Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

```python
over_50.head()
```

```python
#Subplot of chol,age,thalach
```

```python
fig, (ax0,ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10),
                           sharex=True)
#Add data to ax0
scatter= ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

#customize ax0
ax0.set(title="Heart disease & Cholestrol Levels report",
        ylabel="Cholestrol")
#Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")

# Add a mean line

ax0.axhline(over_50["chol"].mean(),
           linestyle='--');


# for ax1

scatter= ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

#customize ax1
ax1.set(title="Heart disease & Maxheart-Rate Levels report",
        ylabel="Thalach",
        xlabel="Age");
#Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

# Add a mean line

ax1.axhline(over_50["thalach"].mean(),
           linestyle='--');

#Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16 , fontweight="bold");
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
fig, (ax0,ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10),
                           sharex=True)
#Add data to ax0
scatter= ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                    cmap="winter")

#customize ax0
ax0.set(title="Heart disease & Cholestrol Levels report",
        ylabel="Cholestrol")
#Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")

# Add a mean line

ax0.axhline(over_50["chol"].mean(),
           linestyle='--');

#setting X-lim
ax0.set_xlim([50,80])


# for ax1

scatter= ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                    cmap="autumn")

#customize ax1
ax1.set(title="Heart disease & Maxheart-Rate Levels report",
        ylabel="Thalach",
        xlabel="Age");
#Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

# Add a mean line

ax1.axhline(over_50["thalach"].mean(),
           linestyle='--');

ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

#Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16 , fontweight="bold");
```
